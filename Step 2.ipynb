# Step 2 -- Plot bike-share data with Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import numpy as np
```

```python
%matplotlib inline
```

```python
data = pd.read_table('daily_weather.tsv')
stations = pd.read_table('stations.tsv')
usage = pd.read_table('usage_2012.tsv')
```

###Convert the date column into actual datetimes.

```python
for i, row in enumerate(data['date']):
        data.set_value(i, 'date', datetime.datetime.strptime(row, '%Y-%m-%d').date())
```

### Make a new month column and fill it with the proper month.

```python
for i, row in enumerate(data['date']):
    data.set_value(i, 'month', row.month)
```

##1. Plot the daily temperature over the course of the year. (This should probably be a line chart.)

```python
data.groupby('month')['temp'].plot()
```

## Create a bar chart that shows the average temperature and humidity by month.

```python
data[['temp', 'humidity', 'month']].groupby('month').mean().plot(kind='bar')
```

## 2. Use a scatterplot to show how the daily rental volume varies with temperature. Use a different series (with different colors) for each season.

```python
d = {'Fall': 'orange', 'Winter': '#eeeeee', 'Spring': 'b', 'Summer': 'y'}
for i, j in data[['total_riders', 'temp', 'season_desc']].groupby('season_desc'):
    plt.scatter(j.temp, j.total_riders, c = d[i], label = i, s=80, alpha=.4)
```

## 3. Create another scatterplot to show how daily rental volume varies with windspeed.  As above, use a different series for each season.

```python
d = {'Fall': 'orange', 'Winter': '#eeeeee', 'Spring': 'b', 'Summer': 'y'}
for i, j in data[['total_riders', 'windspeed', 'season_desc']].groupby('season_desc'):
    plt.scatter(j.windspeed, j.total_riders, c = d[i], label = i, s=80, alpha=.4)
    plt.legend(loc='lower right')
```

## 4. How do the rental volumes vary with geography? Compute the average daily rentals for each station and use this as the radius for a scatterplot of each station's latitude and longitude.

```python
locs = stations[['lat', 'long', 'station']]
avr = pd.concat([usage['station_start']], axis=1, keys=['station'])
count = avr['station'].value_counts()
temp = pd.DataFrame(count.index, columns=['station'])
temp['total_riders'] = count.values
temp['avr'] = temp['total_riders'].apply(lambda c: c/366.0)

locs_and_avr = pd.merge(locs, temp, left_on='station', right_on='station')

plt.scatter(locs_and_avr['long'], locs_and_avr['lat'], c='r',
            s=(locs_and_avr['avr'] * 5), alpha=.3)
```
